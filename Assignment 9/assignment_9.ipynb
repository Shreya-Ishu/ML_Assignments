```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
data = [df1,df2]
df = pd.concat(data)
```

```python
df['Survived'].isnull().sum()
```

```python
df['Survived'] = df['Survived'].dropna()
df['Survived'].isnull().sum()
```

```python
df["Has_Cabin"] = df["Cabin"].isnull()
```

```python
df.head()
```

```python
import re
pattern = re.compile('(\w*[.])')
title_list = []
titles = ["Mr.", "Mrs.", "Master.", "Miss."]
for i in df['Name']:
    title = pattern.findall(i)
    title = ''.join(title) #converting list type to str
    if title in titles:
        title_list.append(title[0:-1])
    else:
        title_list.append('Special')
    
df['Title'] = title_list   
```

```python
df.describe(include="all")
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
df['Embarked'] = df['Embarked'].fillna('S')
df['Cabin'] = df['Cabin'].fillna('C25')
```

```python
df['Categorical_age'] = pd.qcut(df['Age'],4,labels =False)
df['Categorical_fare'] = pd.qcut(df['Fare'],4,labels =False)
```

```python
df = df.drop(['Cabin','Name','PassengerId','Ticket','Age','Fare','SibSp','Parch'],1)
```

```python
df=pd.get_dummies(df)
```

```python
df.head()
```

```python
df.columns
```

```python
x = df[['Embarked_C', 'Embarked_Q','Pclass','Sex_female', 'Sex_male','Has_Cabin','Categorical_age','Categorical_fare','Title_Master','Title_Miss','Title_Mr','Title_Mrs','Title_Special']]
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train,y_train)
```

```python
predictions = regressor.predict(x_test)
```

```python
regressor.score(x_test,y_test)
```
