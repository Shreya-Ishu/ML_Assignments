```python
import numpy as np
a=np.random.rand(10,1)
b=np.mean(a)
print("Mean",b)
```

```python
import numpy as np
a=np.random.rand(20,1)
b=np.std(a)
print("standard deviation",b)
c=(b)**2
print("variance",c)
```

```python
import numpy as np
a=np.random.rand(10,20)
b=np.random.rand(20,25)
c=np.matmul(a,b)
x=np.sum(c)
print(c)
print(x)
```
