# Bivariate Analysis

# Scatter Plot

```python
df.plot(kind='scatter', x = 'pH', y = 'alcohol')
```

```python
sns.jointplot(x='pH', y='alcohol', data=df)
```

# Hexbin Plots

```python
 sns.jointplot(x = df['pH'], y =df['alcohol'], kind='hex',  color='R')
```

```python
 sns.jointplot(x = df['residual sugar'], y =df['alcohol'], kind='hex',  color='K')
```

```python
sns.jointplot(x='pH', y='alcohol', data=df, kind='kde')
```

```python
sns.jointplot(x='quality', y='density', data=df, kind='kde')
```

```python
#Barplot
sns.barplot(x='quality', y='density', data=df)
```

```python
#countplot
sns.countplot(x='quality', data=df, palette='Greens_d')
```

```python
#Swarm plot
sns.swarmplot(x ='quality' , y='sulphates', data=df)
```

# Multivariate Analysis

```python
temp_df=df[['pH','quality','fixed acidity','density']]
sns.pairplot(temp_df)
```

```python
temp_df=df[['volatile acidity','citric acid','residual sugar','pH']]
sns.pairplot(temp_df)
```
